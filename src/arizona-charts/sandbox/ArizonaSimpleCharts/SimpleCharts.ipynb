```python
import numpy
import pandas
import matplotlib.pyplot as plot 
#Load the hospitalizzation inside the csv file.
#The data are avaible in data/ArizonaWidespread directory
pathHospitalizzationData = open('Hospitalizations.csv')
hospitalizzationData = pandas.read_csv(pathHospitalizzationData)
inpatients = hospitalizzationData['Inpatients']
date = hospitalizzationData['Date']
```
